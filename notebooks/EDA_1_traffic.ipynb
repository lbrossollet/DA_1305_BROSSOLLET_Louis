```python
import pandas as pd

# Charger les fichiers CSV
df_selected = pd.read_csv('../data_processed/traffic_data_selected.csv')
df_geo = pd.read_csv('../data_processed/traffic_data_geo.csv')
df_quartiers = pd.read_csv('../data_processed/quartier_paris.csv', sep=';')

# Remplacer les espaces par _ dans les noms de colonnes
df_selected.columns = df_selected.columns.str.replace(' ', '_')
df_geo.columns = df_geo.columns.str.replace(' ', '_')
df_quartiers.columns = df_quartiers.columns.str.replace(' ', '_')

# Renommer la colonne Identifiant arc en Identifiant_arc dans df_geo
df_geo.rename(columns={'Identifiant_arc': 'Identifiant_arc'}, inplace=True)

# Fusionner df_selected avec df_geo pour obtenir N_SQ_QU
df_merged = pd.merge(df_selected, df_geo[['Identifiant_arc', 'N_SQ_QU', 'latitude', 'longitude']], on='Identifiant_arc', how='left')

# Fusionner avec df_quartiers pour obtenir l'arrondissement (C_AR)
df_merged = pd.merge(df_merged, df_quartiers[['N_SQ_QU', 'C_AR']], on='N_SQ_QU', how='left')

# Liste des arrondissements à inclure
arrondissements_inclus = [9, 10, 11, 18]

# Filtrer pour inclure seulement les arrondissements spécifiés
df_traffic_filtered = df_merged[df_merged['C_AR'].isin(arrondissements_inclus)]

# Sélectionner les colonnes requises
columns_selected = ['Identifiant_arc', 'Libelle', 'Date_et_heure_de_comptage', 'Débit_horaire', 'Taux_d\'occupation', 'Etat_trafic', 'latitude', 'longitude', 'C_AR']
df_traffic_filtered = df_traffic_filtered[columns_selected]

# Enregistrer le DataFrame filtré en CSV
df_traffic_filtered.to_csv('traffic_data_filtered.csv', index=False)

print("DataFrame traffic_data_filtered créé et enregistré en traffic_data_filtered.csv.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('../data_processed/df_filtered_north_zone.csv')

# Calculer le pourcentage de blocage par quartier
quartier_bloque_stats = df[df['Etat_trafic'] == 'Bloqué'].groupby('Libelle')['Etat_trafic'].count() / df.groupby('Libelle')['Etat_trafic'].count()
quartier_bloque_stats = quartier_bloque_stats.sort_values(ascending=False).head(10)

# Créer un graphique à barres pour les quartiers les plus bloqués
plt.figure(figsize=(12, 6))
sns.barplot(x=quartier_bloque_stats.values, y=quartier_bloque_stats.index, palette='viridis')
plt.xlabel('Pourcentage de blocage')
plt.ylabel('Quartier')
plt.title('Top 10 des quartiers les plus bloqués (% de bloqué)')
plt.show()
```

```python
# Convertir la colonne Date_et_heure_de_comptage en datetime si ce n'est pas déjà fait
df['Date_et_heure_de_comptage'] = pd.to_datetime(df['Date_et_heure_de_comptage'])

# Extraire le jour de la semaine
df['jour_de_semaine'] = df['Date_et_heure_de_comptage'].dt.weekday

# Calculer le pourcentage de blocage par jour de la semaine
jour_bloque_stats = df[df['Etat_trafic'] == 'Bloqué'].groupby('jour_de_semaine')['Etat_trafic'].count() / df.groupby('jour_de_semaine')['Etat_trafic'].count()
jour_bloque_stats.index = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']

# Créer un graphique à barres pour les jours les plus bloqués
plt.figure(figsize=(10, 6))
sns.barplot(x=jour_bloque_stats.index, y=jour_bloque_stats.values, palette='viridis')
plt.xlabel('Jour de la semaine')
plt.ylabel('Pourcentage de blocage')
plt.title('Jours de la semaine les plus bloqués (% de bloqué)')
plt.ylim(0, 0.1)  # Limiter l'axe y de 0 à 1 pour les pourcentages
plt.show()
```

```python
import pandas as pd

# Chargement des données de débit horaire
file_path = '../data_processed/df_filtered_north_zone.csv'
df_traffic = pd.read_csv(file_path, parse_dates=['Date_et_heure_de_comptage'])

# Chargement du fichier des vacances
vacances_file_path = '../data_processed/vacances.csv'
df_vacances = pd.read_csv(vacances_file_path, parse_dates=['start_date', 'end_date'])

# Fusionner les données de débit horaire avec les informations de vacances
df_merged = pd.merge(df_traffic, df_vacances, left_on='Date_et_heure_de_comptage', right_on='start_date', how='left')

# Marquer les enregistrements pendant les vacances
df_merged['vacances'] = 0  # Initialiser la colonne vacances à 0
vacances_filter = (df_merged['Date_et_heure_de_comptage'] >= df_merged['start_date']) & (df_merged['Date_et_heure_de_comptage'] <= df_merged['end_date'])
df_merged.loc[vacances_filter, 'vacances'] = 1  # Marquer les périodes de vacances avec 1

# Sauvegarde du dataframe modifié dans le fichier CSV d'origine
df_merged.to_csv(file_path, index=False)

# Vérification
print("Colonnes après modification :")
print(df_merged.columns)
print("\nExemple de lignes avec la colonne 'vacances' ajoutée :")
print(df_merged.head())

```

```python
# Mapping pour remplacer les valeurs dans la colonne 'Etat_trafic'
etat_mapping = {
    'Fluide': 1,
    'Présaturé': 2,
    'Saturé': 3,
    'Bloqué': 4
}

df_regroup['Etat_trafic'] = df_regroup['Etat_trafic'].map(etat_mapping).fillna(df_regroup['Etat_trafic'])

# Ajout de la colonne 'AM/PM' en fonction de l'heure
df_regroup['AM/PM'] = df_regroup['heure'].apply(lambda x: 'AM' if int(x[:2]) < 13 else 'PM')

# Affichage du dataframe après les modifications
```
